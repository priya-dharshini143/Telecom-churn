```python
#pivot table-> can b applied for more than two funcs
data.pivot_table(['Total day calls','Total eve calls','Total night calls'],['Area code'],aggfunc='mean')
```

```python
total_calls=(data['Total day calls']+data['Total eve calls']+data['Total night calls'])
data.insert(loc=len(data.columns),column='Total calls',value=total_calls)
data.head()
```

```python
data['Total charge']=(data['Total day charge']+data['Total eve charge']+data['Total night charge']+data['Total intl calls'])
data.head()
```

```python
#to remove col
data.drop(['Total charge','Total calls'],axis=1,inplace=True)
data.drop([1,2]).head()
```
